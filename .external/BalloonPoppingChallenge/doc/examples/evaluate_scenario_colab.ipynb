## Section 1: Import Required Libraries

Import necessary libraries for the evaluation.

```python
import numpy as np

from BalloonPoppingGymEnv.agents.base_agent import BaseAgent
from BalloonPoppingGymEnv.evaluation.evaluate import evaluate_scenario
```

## Section 2: Develop Agent

Develop your agent here in this notebook. You can also import your agent from a separate Python file if you prefer.

```python
# from from BalloonPoppingGymEnv.agents.your_agent import YourAgent

class NoActionAgent(BaseAgent):
    """An agent that does nothing"""

    def __init__(self, *args, **kwargs):
        super().__init__(*args)

        self.launch_time = kwargs.get("launch_time", 0.0)

    def get_action(self, observation):
        """compute agent's action given observation

        This function is necessary to define as it overrides
        an abstract method
        """

        if observation["simulation_time"] >= self.launch_time:
            launch = True
        else:
            launch = False

        return {
            "launch": launch,
            "launch_inclination_heading": np.array([90, 180]),
            "tvc": np.array([0, 0]),
            "roll": 0,
            "throttle": 1,
        }
```

## Section 3: Run Evaluation

Evalate the agent.

```python
scenario_number = 0
render_mode = None  # rendering not supported in notebook environment yet
agent_name = "no_action_agent"
agent_kwargs = {"launch_time": 1.0}

# Load agent class dynamically from specified module path.
evaluate_scenario(
    NoActionAgent,
    agent_kwargs=agent_kwargs,
    agent_name=agent_name,
    scenario_number=scenario_number,
    render_mode=render_mode,
)
```
